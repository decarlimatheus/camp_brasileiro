# Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Carregando dados

```python
camp = pd.read_csv('campeoes_brasileiros.csv')
```

```python
camp.shape
```

```python
camp.head()
```

```python
camp.tail()
```

```python
camp.columns
```

```python
camp.dtypes
```

```python
camp[camp.duplicated()]
```

```python
camp['Vice'] = camp['Vice'].replace('Vasco da Gama', 'Vasco')
camp['Terceiro colocado'] = camp['Terceiro colocado'].replace('Vasco da Gama', 'Vasco')
camp['Quarto colocado'] = camp['Quarto colocado'].replace('Vasco da Gama', 'Vasco')
```

# Análise de Campeões Brasileiros de 1959 até 2022

```python
titulos = camp.groupby('Vencedor')['Ano'].count().reset_index()
```

```python
titulos
```

```python
cores = {'Palmeiras': 'green', 'Santos': 'white', 'Flamengo': 'red', 'Corinthians': 'black', 'São Paulo':'red', 'Fluminense': 'maroon', 'Vasco': 'black', 'Cruzeiro': 'blue','Internacional':'red', 'Grêmio': 'skyblue', 'Botafogo':'black', 'Bahia': 'skyblue', 'Atlético Mineiro':'black', 'Atlético Paranaense': 'red', 'Guarani':'green', 'Coritiba': 'green', 'Sport': 'red'}
```

```python
plt.figure(figsize = (16, 6))

sns.barplot(data = titulos.sort_values(by ='Ano', ascending = False), 
            x = 'Vencedor', 
            y = titulos['Ano'],
            palette=cores,
           edgecolor= 'black')

plt.xticks(rotation = 50)
plt.title('Campeões Brasileiros de 1959 até 2022')
plt.xlabel("Clube")
plt.ylabel("Quantidade de Títulos")
plt.show()
```

# Análise de Campeões Brasileiros de 1971 até 2022

```python
camp_71 = camp[(camp['Ano'] >= 1971) & (camp['Ano'] <= 2022)]
camp_71
```

```python
titulos71 = camp_71.groupby('Vencedor')['Ano'].count().reset_index()
```

```python
titulos71
```

```python
plt.figure(figsize = (16, 6))

sns.barplot(data = titulos71.sort_values(by ='Ano', ascending = False), 
            x = 'Vencedor', 
            y = titulos71['Ano'],
            palette= cores,
           edgecolor='black')

plt.xticks(rotation = 50)
plt.title('Campeões Brasileiros de 1971 até 2022')
plt.xlabel("Clube")
plt.ylabel("Quantidade de Títulos")
plt.show()
```

# Análise de Vice-Campeões Brasileiros de 1971 até 2022

```python
cores_vices71 = {'Palmeiras': 'green', 'Santos': 'white', 'Flamengo': 'red', 'Corinthians': 'black', 'São Paulo':'red', 'Fluminense': 'maroon', 'Vasco': 'black', 'Cruzeiro': 'blue','Internacional':'red', 'Grêmio': 'skyblue', 'Botafogo':'black', 'Bahia': 'skyblue', 'Atlético Mineiro':'black', 'Atlético Paranaense': 'red', 'Guarani':'green', 'Coritiba': 'green', 'Sport': 'red', 'São Caetano': 'Blue', 'Bangu':'red', 'Portuguesa':'red', 'Bragantino': 'white', 'Vitória':'red'}
```

# Análise de Participações entre os 4 primeiros colocados do Brasileiro de 1959 até 2022

```python
camp['G4'] = camp['Vencedor'] + ', ' + camp['Vice'] + ', ' + camp['Terceiro colocado'] + ', ' + camp['Quarto colocado']
```

```python
camp.head()
```

```python
contagem_g4 = camp['G4'].str.split(', ').explode().value_counts().reset_index()
```

```python
contagem_g4
```

```python
plt.figure(figsize = (16, 6))

sns.barplot(data = contagem_g4.sort_values(by ='G4', ascending = False), 
            x = 'index', 
            y = contagem_g4['G4'],
            palette='blend:lightblue,navy')

plt.xticks(rotation = 70)
plt.show()
```

```python
cores_g4_03 = {'Palmeiras': 'green', 'Santos': 'white', 'Flamengo': 'red', 'Corinthians': 'black', 'São Paulo':'red', 'Fluminense': 'maroon', 'Vasco': 'black', 'Cruzeiro': 'blue','Internacional':'red', 'Grêmio': 'skyblue', 'Botafogo':'black', 'Bahia': 'skyblue', 'Atlético Mineiro':'black', 'Atlético Paranaense': 'red', 'Guarani':'green', 'Coritiba': 'green', 'Sport': 'red', 'São Caetano': 'Blue', 'Bangu':'red', 'Portuguesa':'red', 'Bragantino': 'white', 'Vitória':'red', 'Goiás': 'green', 'Fortaleza':'blue'}
```

```python
plt.figure(figsize = (16, 6))

sns.barplot(data = contagem_g4_03.sort_values(by ='G4', ascending = False), 
            x = 'index', 
            y = contagem_g4_03['G4'],
            palette=cores_g4_03,
           edgecolor='black')

plt.xticks(rotation = 40)
plt.title('Participações no G4 do Brasileiro de 2003 até 2022 (Pontos Corridos)')
plt.xlabel("Clube")
plt.ylabel("Quantidade de Participações no G4")
plt.show()
```
